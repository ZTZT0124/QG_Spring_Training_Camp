**定义**：给学习算法提供包含正确答案的示例，让模型学会了在没有给出输出结果的情况下仅凭输入就能给出相当准确的预测或猜测结果

### 1. 回归算法

**定义**：在无限多个数字中尝试预测一个具体的数值                                           
**例子**：根据过去几年的房屋价格预测未来的房屋价格
#### *线性回归*
**训练集**：负责训练模型，使模型达到预期的效果，包含输入的变量也包含输出的目标值 

公式：

**预测函数和代价函数：**
$$
f_{w,b}(x) = w_1x_1 + w_2x_2 + w_3x_3 + \dots + w_nx_n + b
$$

$$
J(w,b) = \frac{1}{2m} \sum_{i=1}^m \left( w_1x_1^{(i)} + w_2x_2^{(i)} + \dots + w_nx_n^{(i)} + b - y^{(i)}\right)^2
$$
> 用“均方误差”描述代价，代价最小时对应的w和b就是要找的目标函数的w和b
**误差函数乘1/2的原因：**使误差函数求偏导数之后的形式更加简洁

**梯度下降：**对误差函数J(w,b)求偏导数，并且同时更新每个w和b
> 偏导数：方向 
>
> 学习率$\alpha$：步长，太大容易导致震荡不收敛，太小容易导致收敛速度太慢
>
> 循环更新：走 

$$
\frac{\partial J(w,b)}{\partial w_1} = \frac{1}{m} \sum_{i=1}^m \left( w_1 x_1^{(i)} + w_2 x_2^{(i)} + \dots + w_n x_n^{(i)} + b - y^{(i)} \right) x_1^{(i)}
$$


$$
\frac{\partial J(w,b)}{\partial b} = \frac{1}{m} \sum_{i=1}^m \left( w_1 x_1^{(i)} + w_2 x_2^{(i)} + \dots + w_n x_n^{(i)} + b - y^{(i)} \right) 
$$

$$\dots$$ 

循环更新（同时更新）：
$$
w_1 = w_1 - \alpha \frac{\partial J(w,b)}{\partial w_1}
$$

$$\dots$$ 

$$
b = b - \alpha \frac{\partial J(w,b)}{\partial b}
$$


#### 向量化

- 目的：使模型执行梯度下降时速度更快

- 工具：numpy

多特征时的预测函数：
$$
f_{w,b}(x) = w_1x_1 + w_2x_2 + w_3x_3 + w_4x_4 + \dots + w_nx_n + b
$$形式类似向量点乘的形式


权重向量：
$$
\vec{w} = (w_1, w_2, w_3, w_4, \dots, w_n)
$$

特征向量：
$$
\vec{x} = (x_1, x_2, x_3, x_4, \dots, x_n)
$$

$$
f_{w,b}(x) = \vec{w} \cdot \vec{x} + b
$$

**多元线性回归可以进行向量化：**

代码简洁，容易编写，并且运行速度快

**例子：**


```python
import numpy as np

w = np.array([1.0,2.5,-3.3])
b = 4
x = np.array([10,20,30])

f_wb = np.dot(w,x) + b
print(f_wb)
```

#### 特征缩放

- 目的：让“等高线”分布更加均匀，防止梯度下降时模型在正确结果附近震荡导致速度变慢
- 方法1：每个数据除以训练集中的最大值
- 方法2：均值归一化

举例：

取值范围：
$$
300 \leq x_1 \leq 2000 \quad 0 \leq x_2 \leq 5
$$

均值：
$$
\mu_1 = \bar{x}_1, \quad \mu_2 = \bar{x}_2
$$

归一化公式：
$$
x_1 = \frac{x_1 - \mu_1}{\max - \min}, \quad x_2 = \frac{x_2 - \mu_2}{\max - \min}
$$


- 方法3：Z-score归一化

标准差：$\sigma$

$$
x_1 = \frac{x_1 - \mu_1}{\sigma_1}
$$

$$
x_2 = \frac{x_2 - \mu_2}{\sigma_2}
$$



### 2. 分类算法

**定义**：从小的、有限可能的输出类别中进行分类和输出

**例子**：根据患者的年龄和肿瘤大小，预测肿瘤是良性的或是恶性的

#### 逻辑回归

专门解决二分类问题的算法

**sigmoid函数：**$$g(z) = \frac{1}{1 + e^{-z}}$$

- 核心性质：无论z取任何值，g(z)的输出永远在(0,1)之间
- 理解：输出一个概率

公式：

**定义Loss函数：**
$$L(f_{\vec{w},b}(x^{(i)}) - y^{(i)})$$
=> 用于衡量单个预测结果与样本的误差

于是在上文的线性回归中，Loss函数为：$$J(\vec w,b) = \vec{w} \cdot \vec{x} + b - y^{(i)}$$

对于逻辑回归，Loss函数为：
\begin{cases}
- \log\left(f_{\vec{w},b}(x^{(i)})\right) & y^{(i)} = 1 \\
- \log\left(1 - f_{\vec{w},b}(x^{(i)})\right) & y^{(i)} = 0
\end{cases}

- 如果真实是 1，模型预测的概率越接近 1，损失越小；越接近 0，损失爆炸式增长

- 如果真实是 1，模型预测的概率越接近 1，损失越小；越接近 0，损失爆炸式增长

将其合并为一个公式可以写为：$$J(\vec{w},b) = -\frac{1}{m}\sum_{i=1}^m \left[ y^{(i)} \log\left(f_{\vec{w},b}(x^{(i)})\right) + (1-y^{(i)}) \log\left(1-f_{\vec{w},b}(x^{(i)})\right) \right]$$

**梯度下降：**逻辑回归的梯度下降，和线性回归的形式完全一致，唯一的区别是$f_{\vec w, b}(x^{(i)})$的定义不同（线性回归是线性输出，逻辑回归是 Sigmoid 后的概率输出）

### 3. 线性回归和逻辑回归的核心特点及异同对比

#### 3.1 核心特点

线性回归的核心特点：

1. 专门解决连续值预测问题（比如房价、气温、销量）
2. 每个$w_j$的大小代表对应特征对结果的影响程度,类似于权重
3. 代价函数是凸函数，梯度下降一定能找到全局最优解
4. 对异常值非常敏感，一个极端异常点会严重拉偏拟合结果

逻辑回归的核心特点：

1. 专门解决二分类问题，输出 0-1 之间的值可解释为概率
2. $w_j$代表对应特征对分类结果的影响程度
3. 代价函数是凸函数，梯度下降可以找到全局最优解
4. 对异常值的敏感度远低于线性回归

#### 3.2 异同对比


| 对比维度       | 线性回归                                       | 逻辑回归                                               |
|----------------|------------------------------------------------|--------------------------------------------------------|
| 核心用途       | 预测连续数值（回归任务）                       | 解决二分类问题（分类任务）                             |
| 模型输出       | $(-\infty, +\infty)$ 的连续值                 | $(0, 1)$ 之间的概率值                                 |
| 核心函数       | 直接使用线性函数 $w\cdot x+b$                  | 线性函数 + Sigmoid 的复合函数                            |
| 代价函数       | 均方误差（MSE）                                | 对数似然                                      |
| 异常值敏感度   | 极高，异常点会严重影响拟合结果                 | 较低，对异常值鲁棒性更好                               |
| 核心假设       | 特征与标签呈线性关系         | 样本服从伯努利分布        |

### 4. 模型评估

- **拟合：**

    定义：模型通过学习训练数据，调整自身参数，尽可能准确地捕捉输入特征与输出标签之间潜在规律的过程。
    
    类型：过拟合和欠拟合（后面有补充）


- **泛化（模型“举一反三”的能力）：**

    定义：训练好的模型在从未见过的新数据上，仍然能够做出正确预测或有效决策的能力。

    当模型的泛化能力差时，通常对应两种情况：

    过拟合：模型过度拟合训练集的数据，在训练集数据的代价函数极小，但在新数据上表现很差。

    欠拟合：模型太简单，连训练集中数据的基本规律都没有找到，对新数据拟合的表现同样很差。


- **训练集、验证集、测试集:**

    来源：将数据集分为三个部分：训练集（60%）、验证集（20%）、测试集（20%）
    
    训练集：对模型进行训练,使代价函数$J_{train}(\vec v, b)$最终收敛到一个比较小的值
    
    > 当模型的代价函数$J_{train}(\vec v, b)$不收敛或者收敛到一个很大的值时，成为欠拟合

    验证集：根据代价函数$J_{cv}(\vec w,b)$最终收敛的大小对模型进行评估
    
    > 当模型训练集的代价函数$J_{train}(\vec v, b)$收敛到一个很小的值，但是验证集的代价函数$J_{cv}(\vec w,b)$的
    值却很大的时候，成为过拟合，即模型对训练集的数据拟合得很好，但是对新的数据却不能作出正确得预测

    测试集：根据代价函数$J_{test}(\vec w, b)$最终评估模型的泛化能力



## 无监督学习

### 聚类算法

**定义**：处理无标签的数据，找到关键词或相同的结构、模式，并尝试自动将他们分组，拥有相同特征的数据分布在一起

**例子**：为用户推荐他们更可能去点击的广告、依据关键词对新闻进行分组

**K均值算法：**

- 定义：根据每个数据点到每个聚类中心的距离，不断为每个数据点选择离它最近的聚类中心，然后更新每个距离中心的位置。一直迭代直到聚类中心位置不再发生变化

公式：

$$J(c(1),c(2),...,c(m),\mu_1,\mu_2,...,\mu_k) = \frac{1}{m} \sum_{i=1}^{m} \| x^{(i)} - \mu_{c(i)} \|^2$$

- $\| x^{(i)} - \mu_{c(i)} \|^2$：第$i$个样本到其聚类中心的**欧氏距离平方**
- 除以$m$是为了得到平均代价
K-Means的最终优化目标可表示为：
$$\min_{c(1)...c(m), \mu_1...\mu_k} J(c(1),...,c(m),\mu_1,...,\mu_k)$$

#### 优化

##### 第一步： 固定聚类中心，优化样本
保持所有聚类中心不变，对每个样本$x^{(i)}$，将其分配到距离最近的聚类中心：

$$c(i) := \arg\min_{j \in \{1,...,k\}} \| x^{(i)} - \mu_j \|^2$$


##### 第二部：固定样本，优化聚类中心
保持所有样本的聚类标签不变，计算每个聚类中所有样本的均值，将均值作为新的聚类中心：

$$\mu_j := \frac{1}{m_j} \sum_{i: c(i)=j} x^{(i)}$$

- $m_j$：第$j$个簇的样本数量

#### 迭代
1. **初始化**：随机选择$k$个样本作为初始聚类中心
2. **迭代直至收敛**：重复步骤1和步骤2，直到代价函数/聚类中心变化小于阈值$\epsilon$：
   - 第$t$次迭代的簇分配：
     $$c^{(t)}(i) = \arg\min_{j} \| x^{(i)} - \mu_j^{(t-1)} \|^2$$
   - 第$t$次迭代的中心更新：
     $$\mu_j^{(t)} = \frac{1}{m_j^{(t)}} \sum_{i: c^{(t)}(i)=j} x^{(i)}$$
3. **收敛条件**：
   - 代价函数收敛：$|J^{(t)} - J^{(t-1)}| < \epsilon$；
   - 中心收敛：$\sum_{j=1}^k \| \mu_j^{(t)} - \mu_j^{(t-1)} \|^2 < \epsilon$（$\epsilon$通常取$10^{-6} \sim 10^{-9}$）。


## PCA降维技术
##### 定义：
**降维技术**，将高维数据映射到低维空间，同时保留数据中的方差特征，实现维度压缩。

#### 维度灾难
1. 维度灾难的核心问题
    太高维度的数据会导致：
    - 过拟合风险升高
    - 运行太慢
    
#### PCA的核心作用
1. 维度压缩
    - 降低计算成本，减少特征维度，加速模型训练
    - 简化数据表示

2. 数据去噪与特征提纯，可以去除冗余的特征

3. 数据可视化
    - 将高维数据（如100维）压缩至2/3维，可直观绘制散点图，分析样本分布、聚类趋势、异常值等；
 
4. 可减少模型对冗余特征的学习，降低过拟合风险

#### PCA降维的关键原则
1. 主成分选择：保留的主成分数量需满足“累计方差贡献率”阈值，降低维度的同时不丢失关键的信息；
2. 无监督：仅从数据本身的方差结构出发降维，适用于所有类型的高维数据预处理；
